```python
confirmed = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
recovered = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
deaths = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
```

```python
def create_dataset(df):

    df = pd.melt(df, 
            id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
            var_name='date', 
            value_name='value')
    df['date'] = pd.to_datetime(df['date'])
    df = df.rename(columns={'Country/Region':'Country'})
    df = df.drop(columns=['Province/State', 'Lat', 'Long'])
    df = df.sort_values(by=['Country', 'date'], ascending=[True, True])
    
    return df
```

```python
confirmed = create_dataset(confirmed)
deaths = create_dataset(deaths)
recovered = create_dataset(recovered)
```

```python
df = pd.merge(confirmed, deaths, on=['Country', 'date'], suffixes=['_confirmed', '_deaths'], how='left')
df = pd.merge(df, recovered, on=['Country', 'date'], how='left')
df = df.rename(columns={'value':'value_recovered'})
```

```python
df['week'] = df['date'].dt.to_period('W-MON') 
```

```python
weekly_df = df \
   .groupby(['Country', 'week']) \
   .agg({'value_confirmed':'sum', 'value_deaths':'sum', 'value_recovered':'sum'}) \
   .reset_index()
```

```python
weekly_df
```

```python
weekly_df[['value_confirmed', 'value_deaths', 'value_recovered']].corr(method='pearson')
```
