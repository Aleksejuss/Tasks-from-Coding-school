### Perašykite apačioje SQL pateiktas užklausas su Pandas

```python
import pandas as pd
titanic = pd.read_csv('data/titanic.csv')

### titanic stulpelių paaiškinimai:

# survived - Ar išgyveno (0 = Ne; 1 = Taip).
# pclass - Keleivio kelionės klasė (1 = pirma; 2 = antra; 3 = trečia).
# sex - Lytis.
# age - Amžius.
# sibsp - Brolių ir seserų/sutuoktinių skaičius laive .
# parch - Tėvų/vaikų skaičius laive.
# fare - Kaina.
# embarked - Išvykimo uostas (C = Cherbourg; Q = Queenstown; S = Southampton).
# deck - Denis.
# kiti: Perdaryti iš jau minėtų viršuje.
```

```python
# SELECT * FROM titanic LIMIT 10
titanic.head(10)
```

```python
titanic.info()
```

```python
titanic[(titanic['pclass']=="null")]
```

```python
titanic[(titanic['pclass']==titanic['pclass'].isnull().sum())]
```

```python
titanic[(titanic['pclass'].isnull())]
```

```python
# SELECT survived, age FROM titanic 
titanic[['survived','age']]
```

```python
titanic['survived'].drop_duplicates()
```

```python
# SELECT DISTINCT survived FROM titanic 
titanic['survived'].unique()
```

```python
# SELECT AVG(age) FROM titanic
titanic['age'].mean()
```

```python
# SELECT SUM(age) FROM titanic -
titanic['age'].sum()
```

```python
# SELECT * FROM titanic WHERE age BETWEEN 10 AND 20 
```

```python
titanic[(titanic['age']>=10) & (titanic['age']<=20)]
```

```python
titanic[titanic['age'].between(10, 20)]
```

```python
# SELECT * FROM titanic WHERE class LIKE 'Th%
titanic[titanic['class'].str.startswith('Th')]
```

```python
# SELECT * FROM titanic WHERE class LIKE '%rd'
titanic[titanic['class'].str.endswith('rd')]
```

```python
# SELECT * FROM titanic WHERE class LIKE '%ir%'
titanic[titanic['class'].str.contains('ir')]
```

```python
# SELECT * FROM df_a a LEFT JOIN df_b b ON a.subject_id = b.subject_id 
```

```python
# SELECT * FROM titanic ORDER BY age
titanic.sort_values(by=['age'], ascending=[True])
```

```python
# SELECT * FROM titanic ORDER BY age, fare DESC
titanic.sort_values(by = ['age', 'fare'], ascending=[True, False])
```
