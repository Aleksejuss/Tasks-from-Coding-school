```python
import numpy as np
import pandas as pd
```

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')
titanic
```

```python
titanic.info()
```

```python
titanic.isnull().sum()
```

```python
average_values = titanic.groupby('sex')['age'].transform('mean')
average_values
```

```python
titanic['age'] = titanic['age'].fillna(average_values)
```

```python
titanic.isnull().sum()
```

```python
titanic.deck.unique()
```

```python
titanic['deck'] = titanic['deck'].fillna('no_deck')
```

```python
titanic = titanic.dropna()
```

```python
titanic.info()
```

```python
titanic.isnull().sum()
```

```python
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')

np.random.seed(123)
rand_idx = np.random.randint(len(iris), size=int(len(iris)*0.3))
iris.iloc[rand_idx, range(0, 4)] = np.nan

iris
# iris:

# sepal_length: sepal dalies ilgis
# sepal_width: sepal dalies plotis
# petal_length: petal dalies ilgis
# petal_width: petal dalies plotis
# species: gelės rūšis

# https://en.wikipedia.org/wiki/Iris_(plant)    
```

```python
iris.isnull().sum()
```

```python
a = iris.sepal_length.mean()
b = iris.sepal_width.mean()
c = iris.petal_length.mean()
d = iris.petal_width.mean()
```

```python
iris['sepal_length'] = iris['sepal_length'].fillna(a)
iris['sepal_width'] = iris['sepal_width'].fillna(b)
iris['petal_length'] = iris['petal_length'].fillna(c)
iris['petal_width'] = iris['petal_width'].fillna(d)
```

```python
iris
```

```python
property_data = pd.read_csv('https://raw.githubusercontent.com/dataoptimal/posts/master/data%20cleaning%20with%20python%20and%20pandas/property%20data.csv')
property_data = property_data.drop(columns=['PID'])
property_data
```

```python
property_data.info()
```

```python
property_data['ST_NUM'] = property_data['ST_NUM'].fillna(0)
property_data['SQ_FT'] = property_data['SQ_FT'].replace('--',0)
property_data['SQ_FT'] = property_data['SQ_FT'].fillna(0)
property_data['NUM_BATH'] = property_data['NUM_BATH'].fillna(0)
property_data['NUM_BATH'] = property_data['NUM_BATH'].replace('HURLEY',0)
property_data['NUM_BEDROOMS'] = property_data['NUM_BEDROOMS'].fillna(0)
property_data['NUM_BEDROOMS'] = property_data['NUM_BEDROOMS'].replace('na',0)
property_data['OWN_OCCUPIED'] = property_data['OWN_OCCUPIED'].fillna('N')
property_data['OWN_OCCUPIED'] = property_data['OWN_OCCUPIED'].replace('12','Y')
```

```python
property_data
```

```python
property_data.isnull().sum()
```
