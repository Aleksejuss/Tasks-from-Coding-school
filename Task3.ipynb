## 20 Užduotis

```python
vyras = titanic[titanic['lytis'] == "vyras"].mean()['age']
moteris = titanic[titanic['lytis'] == "moteris"].mean()['age']

print(vyras)
print(moteris)
```

```python
if [titanic['lytis'] == "vyras"]:
    titanic['modifikuotas_age2'] = titanic['age'].fillna(vyras)
else:
    titanic['modifikuotas_age2'] = titanic['age'].fillna(moteris)
    
titanic
```

## 22 Užduotis

Apskaičiuokite vidutinį keleivių amžių, standartinį keleivių amžiaus vidurkį. Sukurkite naują stulpelį titanic duomenų lentelėje, kuriame būtų suskaičiuota:
(keleivio amžius - vidutinis keleivio amžius)/keleivio amziaus standartinis nuokrypis.

```python
avg_age = titanic['modifikuotas_age2'].mean()
print(avg_age)
std_age = titanic['modifikuotas_age2'].std()
print(std_age)

titanic['age_normalizacija'] = (titanic['modifikuotas_age2'] - avg_age)/std_age

titanic
```
